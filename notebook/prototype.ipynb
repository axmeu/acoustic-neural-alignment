```python
import pandas as pd
```

```python
df = pd.read_csv("../outputs/table.csv")

print("=== General overview ===")
print(f"Speakers:           {df['speaker_id'].nunique()}")
print(f"Sentences:          {df['sentence_id'].nunique()}")
print(f"WAV files:          {df['wav_path'].nunique()}")
print(f"Phonemes:           {len(df)}")
print(f"Unique phonemes:    {df['phoneme'].nunique()}")

print("\n=== 2x2 design ===")
print("Speakers per cell:")
print(df.groupby(["l1_status", "gender"])["speaker_id"].nunique().unstack())
print("\nTokens per cell:")
print(df.groupby(["l1_status", "gender"]).size().unstack())


print("\n=== Phoneme durations (ms) ===")
print(df["duration_ms"].describe())
```
