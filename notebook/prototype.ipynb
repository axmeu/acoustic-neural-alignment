```python
import parselmouth
from parselmouth.praat import call
import numpy as np
from pathlib import Path

data_dir = Path("/home/onyxia/work/data/ru-fr_interference/2/wav_et_textgrids")
wav = next(data_dir.rglob("*.wav"))
print(wav)
sound = parselmouth.Sound(str(wav))
segment = sound.extract_part(from_time=0.1, to_time=0.3, preserve_times=True)
pitch = call(segment, "To Pitch (ac)",
             0.0,    # time step (0 = auto)
             75.0,   # pitch floor (Hz)
             15,     # max number of candidates
             "yes",  # very accurate
             0.03,   # silence threshold
             0.45,   # voicing threshold
             0.01,   # octave cost
             0.35,   # octave-jump cost
             0.14,   # voiced/unvoiced cost
             600.0)  # pitch ceiling (Hz)
val = call(pitch, "Get value at time", 0.2, "Hertz", "Linear")
print(val)
```
